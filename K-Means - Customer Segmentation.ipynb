```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
df = pd.read_csv("income.csv")
```

```python
df.head()
```

```python
plt.scatter(df['Age'],df['Income($)'])
plt.show()
```

```python
km = KMeans (n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'], color = 'green')
plt.scatter(df2.Age,df2['Income($)'], color = 'Orange')
plt.scatter(df3.Age,df3['Income($)'], color = 'violet')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
plt.show()
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
df

scaler.fit(df[['Age']])
df.Age = scaler.transform(df[['Age']])
df
```

```python
km = KMeans (n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.drop('cluster', axis='columns', inplace = True)
df
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
df['cluster'] = kmeans.fit_predict(df[['Age', 'Income($)']])
```

```python
import matplotlib.pyplot as plt

df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]

plt.scatter(df1.Age, df1['Income($)'], color='red', label='Cluster 0')
plt.scatter(df2.Age, df2['Income($)'], color='orange', label='Cluster 1')
plt.scatter(df3.Age, df3['Income($)'], color='violet', label='Cluster 2')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
plt.show()
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans (n_clusters=k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
plt.show()
```
